```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps as cm

anomaly_data = np.genfromtxt("Land_and_Ocean_summary.txt", comments="%")

year = np.array([elm[0] for elm in anomaly_data])
above = np.array([elm[1] for elm in anomaly_data])
below = np.array([elm[5] for elm in anomaly_data])
```

```python
fig, allyears = plt.subplots()
plt.plot(year, above, color='red', label="Above ice")
plt.plot(year, below, color='blue', label="Below ice")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.legend()
plt.title("Global Below and Above Ice relative Mean")
plt.show()
```

```python
ten_year_indices = np.where(year % 10 == 0)
ten_year_rows = anomaly_data[ten_year_indices]
ten_year_x = year[ten_year_indices]
ten_year_above = above[ten_year_indices]
ten_year_below = below[ten_year_indices]
```

```python
fig, tenyears = plt.subplots()
plt.plot(ten_year_x, ten_year_above, color='red', label="Above ice")
plt.plot(ten_year_x, ten_year_below, color='blue', label="Below ice")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.legend()
plt.title("Global Below and Above Ice: Decade Mean 1850-2018")
plt.show()
```

```python
my_cmap = cm['hsv']
my_cmap

fig, color_bars = plt.subplots()
plt.bar(year, above, color=[my_cmap(i) for i in range(len(year))], width=1)
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.title("Global Surface Temperature relative Mean")
plt.show()
```
